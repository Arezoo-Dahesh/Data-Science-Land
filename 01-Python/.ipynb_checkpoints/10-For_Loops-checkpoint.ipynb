## For loop

```python
# for counter in sequence:
#     statement_1
#     statement_2
#     statement_3
```

## For in string

```python
for x in "Python":
    print(x)
```

## For in list

```python
names = ["Arezoo", "Zahra", "Ehsan", "Hoori", "Ali", "Reza"]
for x in names:
    print(x)
```

### The break Statement

```python
names = ["Arezoo", "Zahra", "Ehsan", "Hoori", "Ali", "Reza"]
for x in names:
    print(x)
    if x == "Ehsan":
        break
```

```python
for x in names:
    if x == "Ehsan":
        break
    print(x)
```

### The continue Statement

```python
names = ["Arezoo", "Zahra", "Ehsan", "Hoori", "Ali", "Reza"]
for x in names:
    if x == "Ehsan":
        continue
    print(x)
```

### The range() Function

```python
# range(start,end,step)
```

```python
for x in range(6):
    print(x)
```

```python
for x in range(2,7):
    print(x)
```

```python
for x in range(0,10,3):
    print(x)
```

```python
print(names)
```

```python
for i in range(len(names)):
    print(i) 
```

```python
for i in range(len(names)):
    print(names[i]) 
```
